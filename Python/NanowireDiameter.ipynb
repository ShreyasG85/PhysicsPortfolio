```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
mask_sizes=[160,240,320] 
n=30
```

```python
diam_files = ['Diameter_Sara_0.csv', 'Diameter_Sara_1.csv', 'Diameter_Sara_2.csv']
height_files = ['Height_Sara_0.csv', 'Height_Sara_1.csv', 'Height_Sara_2.csv']

results = {
    'diam_mean': [], 'diam_std': [], 'diam_ci': [],
    'height_mean': [], 'height_std': [], 'height_ci': []
}

def calc_ci(std_dev, n):
    return 1.96 * (std_dev / np.sqrt(n))

for file in diam_files:
    df = pd.read_csv(file)
    mean = df['Height'].mean()
    std = df['Height'].std()
    results['diam_mean'].append(mean)
    results['diam_std'].append(std)
    results['diam_ci'].append(calc_ci(std, n))

for file in height_files:
    df = pd.read_csv(file)
    mean = df['Height'].mean()
    std = df['Height'].std()
    results['height_mean'].append(mean)
    results['height_std'].append(std)
    results['height_ci'].append(calc_ci(std, n))

# Create Summary DataFrame
summary_df = pd.DataFrame({
    'Mask Opening (nm)': mask_sizes,
    'Mean Diameter (um)': results['diam_mean'],
    'CI Diameter 95%': results['diam_ci'],  
    'Mean Height (um)': results['height_mean'],
    'CI Height 95%': results['height_ci']   
})

print(summary_df)

```

```python
plt.figure(figsize=(8, 6))
plt.errorbar(mask_sizes, results['diam_mean'], yerr=results['diam_ci'],fmt='o-', capsize=5, color='blue', label='mean diameter (95% CI)')
plt.xlabel('mask opening diameter (nm)',fontsize=15)
plt.ylabel('nanowire diameter (um)',fontsize=15)
plt.title('nanowire diameter vs mask opening',fontsize=15)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()

```

```python
plt.figure(figsize=(8, 6))
plt.errorbar(mask_sizes, results['height_mean'], yerr=results['height_ci'],fmt='o-', capsize=5, color='red', label='mean height (95% CI)')
plt.xlabel('mask opening diameter (nm)',fontsize=15)
plt.ylabel('nanowire height (um)',fontsize=15)
plt.title('nanowire height vs mask opening',fontsize=15)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```
