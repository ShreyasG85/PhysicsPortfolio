```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.signal import find_peaks
```

```python
sim_data1 = pd.read_csv('W1_2.csv', skiprows=8)
sim_freq1 = sim_data1['Frequency (GHz)'] * 1e9 
sim_real1 = sim_data1['RE[S11]']
sim_imag1 = sim_data1['IM[S11]']
sim_mag1 = np.sqrt(sim_real1**2 + sim_imag1**2)
sim_mag_dB1 = 20 * np.log10(sim_mag1)
peks1, _ = find_peaks(-sim_mag_dB1, height=-20, distance=5000)
```

```python
sim_data2 = pd.read_csv('W1_3.csv', skiprows=8)
sim_freq2 = sim_data2['Frequency (GHz)'] * 1e9  
sim_real2 = sim_data2['RE[S11]']
sim_imag2 = sim_data2['IM[S11]']
sim_mag2 = np.sqrt(sim_real2**2 + sim_imag2**2)
sim_mag_dB2 = 20 * np.log10(sim_mag2)
peks2, _ = find_peaks(-sim_mag_dB2, height=-20, distance=3000)
```

```python
sim_data3 = pd.read_csv('W1_4.csv', skiprows=8)
sim_freq3 = sim_data3['Frequency (GHz)'] * 1e9  
sim_real3 = sim_data3['RE[S11]']
sim_imag3 = sim_data3['IM[S11]']
sim_mag3 = np.sqrt(sim_real3**2 + sim_imag3**2)
sim_mag_dB3 = 20 * np.log10(sim_mag3)
peks3, _ = find_peaks(-sim_mag_dB3, height=-20, distance=3000)
```

```python
sim_data = pd.read_csv('W1_test_13.csv', skiprows=54)
sim_freq = sim_data['Frequency (GHz)'] * 1e9  
sim_real = sim_data['RE[S11]']
sim_imag = sim_data['IM[S11]']
sim_mag = np.sqrt(sim_real**2 + sim_imag**2)
sim_mag_dB = 20 * np.log10(sim_mag)
peks, _ = find_peaks(-sim_mag_dB, height=-20, distance=3000)
```

```python
sim_data4 = pd.read_csv('W1_5.csv', skiprows=8)
sim_freq4 = sim_data4['Frequency (GHz)'] * 1e9  
sim_real4 = sim_data4['RE[S11]']
sim_imag4 = sim_data4['IM[S11]']
sim_mag4 = np.sqrt(sim_real4**2 + sim_imag4**2)
sim_mag_dB4 = 20 * np.log10(sim_mag4)
peks4, _ = find_peaks(-sim_mag_dB4, height=-20, distance=3000)
```

```python
sim_data5 = pd.read_csv('W1_6.csv', skiprows=8)
sim_freq5 = sim_data5['Frequency (GHz)'] * 1e9  
sim_real5 = sim_data5['RE[S11]']
sim_imag5 = sim_data5['IM[S11]']
sim_mag5 = np.sqrt(sim_real5**2 + sim_imag5**2)
sim_mag_dB5 = 20 * np.log10(sim_mag5)
peks5, _ = find_peaks(-sim_mag_dB5, height=-20, distance=3000)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(sim_freq, sim_mag_dB, label='0 nH (Only Reflection Resonator)', color='green')
plt.plot(sim_freq3, sim_mag_dB3, label='0.5 nH', color='blue')
plt.plot(sim_freq1, sim_mag_dB1, label='1 nH', color='orange')
plt.plot(sim_freq2, sim_mag_dB2, label='5 nH', color='red')
plt.plot(sim_freq4, sim_mag_dB4, label='10 nH', color='black')
plt.plot(sim_freq5, sim_mag_dB5, label='50 nH', color='brown')
plt.xlabel('Frequency (10 GHz)')
plt.ylabel('Magnitude (dB)')
plt.title('S11 Magnitude (dB) vs Frequency')
for pek in peks:
    plt.annotate(f'{sim_freq[pek]/1e9:.2f} GHz', 
                 (sim_freq[pek], sim_mag_dB[pek]),
                 textcoords="offset points", xytext=(0,10), ha='center', color='red')
for pek in peks1:
    plt.annotate(f'{sim_freq1[pek]/1e9:.2f} GHz', 
                 (sim_freq1[pek], sim_mag_dB1[pek]),
                 textcoords="offset points", xytext=(0,10), ha='center', color='red')

for pek in peks2:
    plt.annotate(f'{sim_freq2[pek]/1e9:.2f} GHz', 
                 (sim_freq2[pek], sim_mag_dB2[pek]),
                 textcoords="offset points", xytext=(0,10), ha='center', color='red')
    
for pek in peks3:
    plt.annotate(f'{sim_freq3[pek]/1e9:.2f} GHz', 
                 (sim_freq3[pek], sim_mag_dB3[pek]),
                 textcoords="offset points", xytext=(0,10), ha='center', color='red')
    
for pek in peks4:
    plt.annotate(f'{sim_freq4[pek]/1e9:.2f} GHz', 
                 (sim_freq4[pek], sim_mag_dB4[pek]),
                 textcoords="offset points", xytext=(0,10), ha='center', color='red')
    
for pek in peks5:
    plt.annotate(f'{sim_freq5[pek]/1e9:.2f} GHz', 
                 (sim_freq5[pek], sim_mag_dB5[pek]),
                 textcoords="offset points", xytext=(0,10), ha='center', color='red')
    
plt.grid(True)
plt.legend()
plt.show()
```
