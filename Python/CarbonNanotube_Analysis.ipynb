```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
```

```python
df=pd.read_csv('59% T-350 Ohms per square.csv',skiprows=1)
wavelength=df['Wavelength (nm)']
absorbance=df['Abs']
peaks,properties=find_peaks(absorbance,height=0.2,distance=350)
plt.plot(wavelength,absorbance,label='SWCNT Film')
plt.plot(wavelength[peaks],absorbance[peaks],'r^',label='peak')
plt.xlabel('wavelength(nm)')
plt.ylabel('absorbance')
plt.title('59% T-350 absorption spectrum')
plt.legend()
plt.show()
print("wavelength(nm):",wavelength.iloc[peaks].values)
print("height(Abs):",absorbance.iloc[peaks].values)
```

```python
df = pd.read_csv('SWCNTs lab course- 514nm laser source.txt', 
                 header=None, delimiter='\t', names=['Shift','Intensity'])
plt.plot(df['Shift'],df['Intensity'],label='raman')
peaks,_=find_peaks(df['Intensity'],height=1000,distance=100) 
plt.plot(df['Shift'].iloc[peaks],df['Intensity'].iloc[peaks],
         'r^',label='peaks')
shift=df['Shift'].iloc[peaks]
intensities=df['Intensity'].iloc[peaks]
for x,y in zip(shift,intensities):
    print(f"shift value= {x:.4f}cm^-1, intensity= {y:.4f}")
plt.xlabel('raman shift (cm)$^-1$')
plt.ylabel('intensity (a.u.)')
plt.title('raman spectrum')
plt.legend()
plt.show()
```
