```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
sim_data = pd.read_csv('Filter8.csv', skiprows=8)
sim_freq = sim_data['Frequency (GHz)'] * 1e9  
sim_real = sim_data['RE[S21]']
sim_imag = sim_data['IM[S21]']
sim_mag = np.sqrt(sim_real**2 + sim_imag**2)
sim_mag_dB = 20 * np.log10(sim_mag)
plt.figure(figsize=(10, 6))
plt.plot(sim_freq, sim_mag_dB, label='Simulation S21(dB)', color='blue')
plt.xlabel('Frequency (Hz) [Logarithmic Scale]')
plt.ylabel('Magnitude (dB)')
plt.title('S21 Magnitude (dB) vs Frequency')
plt.xscale('log')
plt.axhline(y=-3, color='red', linestyle='--', label='-3 dB')
plt.axvline(x=4e9, color='green', linestyle='--', label='4 GHz')
plt.axvline(x=8e9, color='purple', linestyle='--', label='8 GHz')
plt.grid(True)
plt.legend()
plt.show()
```
